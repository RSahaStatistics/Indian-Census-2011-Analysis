# Calculate state-wise:

How to delete columns from data

```python
df = data.drop(['District_code','State_name','District_name'],axis=1)
df
```

Describe the data

```python
df.describe()
```

Total population

```python
data_with_Total_population=data.groupby('State_name').Population.sum().sum()
data_with_Total_population
```

Total population statewise

```python
data_with_population=data.groupby('State_name').Population.sum().sort_values(ascending = False)
data_with_population
```

Statewise number of workers

```python
data_with_workers=data.groupby('State_name').Workers.sum().sort_values(ascending = False)
data_with_workers
```

Religious demography

```python
data_religious = data.groupby('State_name')['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains'].sum().sort_values(by='Hindus')
data_religious
```

Educational demography

```python
data_education = data.groupby('State_name')['Secondary_Education','Higher_Education','Graduate_Education'].sum().sort_values(by='Secondary_Education')
data_education
```

Note: Also there is some interesting data in states and also between the Secondary_Education, Higher_Education, Graduate_Education numbers

Calculate statewise literacy

```python
data_with_Literate=data.groupby('State_name').Literate.sum().sort_values(ascending = False)
data_with_Literate
```

Calculate statewise literacy rate

```python
data_with_Literate_percentage=data.groupby('State_name').Literate.sum()*100/data_with_population.sort_values(ascending = False)
data_with_Literate_percentage
```

# Particularly For West Bengal

Access all data of West Bengal

```python
West_Bengal_Data = data[data.State_name == 'WEST BENGAL']
West_Bengal_Data
```

District wise population distribution

```python
data_with_population_West_Bengal = West_Bengal_Data.groupby('District_name').Population.sum().sort_values(ascending = False)
data_with_population_West_Bengal
```

Total Number of Male Workers

```python
data[data.State_name == 'WEST BENGAL']['Male_Workers'].sum()
```

West Bengal religious data

```python
West_Bengal_religious_data = West_Bengal_Data.groupby('District_name')['Hindus','Muslims','Christians','Sikhs','Buddhists','Jains'].sum().sort_values(by='Hindus')
West_Bengal_religious_data
```

West Bengal education data

```python
West_Bengal_education = West_Bengal_Data.groupby('District_name')['Secondary_Education','Higher_Education','Graduate_Education'].sum().sort_values(by='Secondary_Education')
West_Bengal_education
```
